### As the null value percentage for 'Accommodates' columns is 0.046010 which is very small, imputing using the mean value of the same column

```python
ml_airbnb['Accommodates'] = ml_airbnb['Accommodates'].fillna(np.round(ml_airbnb['Accommodates'].mean()))
```

### Taking a look at the heatmap and infering the below points for imputation:
    
     1)The 'Bedrooms' column is highly correlated with 'Accomodates' column. So group bedrooms by accomodates and impute null values with MODE imputation

     2)'Beds' and 'Bathrooms' columns are highly correlated with 'Bedrooms'. So the same but group by bedrooms.


```python
ml_airbnb["Bedrooms"] = ml_airbnb.groupby("Accommodates")['Bedrooms'].transform(lambda x: x.fillna(x.mode()[0]))
ml_airbnb['Beds'] =ml_airbnb.groupby('Bedrooms')['Beds'].transform(lambda x: x.fillna(x.mode()[0]))
ml_airbnb['Bathrooms'] =ml_airbnb.groupby('Bedrooms')['Bathrooms'].transform(lambda x: x.fillna(x.mode()[0]))
```

### The 'Neighbourhood Cleansed' column contains the neighborhoods and since 'State' can be considered as a group of neighborhoods, groupby 'State' and impute the mode values of each state to its own null values.

```python
ml_airbnb['State'] = ml_airbnb.groupby('Neighbourhood Cleansed')['State'].transform(lambda x: x.fillna(x.mode()[0]))
```

### Using the now imputed 'Neighbourhood Cleansed' column,taking the help of 'Room Type' column to determine the 'Property Type' column. Using MODE imputation for the same.

### Using 'Bathrooms', 'Bedrooms', 'Accommodates' and 'Beds' columns to impute 'Cleaning Fee' as Cleaning Fee is highly correlated with the features mentioned before. Imputation done using KNN Imputation

```python
cleaning_fee_df = pd.DataFrame()
cleaning_fee_df['bathrooms'] = ml_airbnb['Bathrooms']
cleaning_fee_df['bedrooms'] = ml_airbnb['Bedrooms']
cleaning_fee_df['Accomodates'] = ml_airbnb['Accommodates']
cleaning_fee_df['Beds'] = ml_airbnb['Beds']
cleaning_fee_df['Cleaning Fee'] = ml_airbnb['Cleaning Fee']

from sklearn.impute import KNNImputer
impu = KNNImputer()
imputed_cleaning = impu.fit_transform(cleaning_fee_df)
cleaning_fee_df_imputed = pd.DataFrame(imputed_cleaning, columns = cleaning_fee_df.columns)
```

```python
cleaning_fee_df_imputed = cleaning_fee_df_imputed.reset_index(drop = True)
ml_airbnb = ml_airbnb.reset_index(drop = True)
ml_airbnb['Cleaning Fee'] = cleaning_fee_df_imputed['Cleaning Fee']
```
